```python
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

from bs4 import BeautifulSoup as bs
import requests

import pandas as pd
import numpy as np

import time

import warnings
warnings.filterwarnings("ignore")

from tqdm.auto import tqdm, trange
```

```python
pages = list(range(1, 402, 50))
# pages = [1, 51, 351, 401]

titles = []
artists = []
albums = []
dates = []
genres = []
flacs = []
lyrics = []


for page in pages:
  url = f"https://www.melon.com/search/song/index.htm?q=%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8&section=artist&searchGnbYn=Y&kkoSpl=Y&kkoDpType=#params%5Bq%5D=%25EB%25B0%25A9%25ED%2583%2584%25EC%2586%258C%25EB%2585%2584%25EB%258B%25A8&params%5Bsort%5D=hit&params%5Bsection%5D=artist&params%5BsectionId%5D=&params%5BgenreDir%5D=&po=pageObj&startIndex={page}"
  print(url)

  driver = webdriver.Chrome("chromedriver", chrome_options=chrome_options)
  # url = "https://www.melon.com/search/song/index.htm?q=%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8&section=artist&searchGnbYn=Y&kkoSpl=Y&kkoDpType=#params%5Bq%5D=%25EB%25B0%25A9%25ED%2583%2584%25EC%2586%258C%25EB%2585%2584%25EB%258B%25A8&params%5Bsort%5D=hit&params%5Bsection%5D=artist&params%5BsectionId%5D=&params%5BgenreDir%5D=&po=pageObj&startIndex=1"
  driver.get(url)

  driver.implicitly_wait(10)

  song_info_list = driver.find_elements_by_css_selector("a.btn.btn_icon_detail") #50개
  # print(len(song_info_list))


  for i in range(len(song_info_list)):    
  # for i in range(4):
    song_info_list[i].send_keys("\n")
    time.sleep(1)
    
    #제목
    if driver.find_element_by_css_selector("div.song_name"):
      title = driver.find_element_by_css_selector("div.song_name")
      titles.append(title.text)
    else:
      titles.append("제목 정보없음")

  
    #가수
    if driver.find_element_by_css_selector("div.artist"):
      artist = driver.find_element_by_css_selector("div.artist")
      artists.append(artist.text)
    else:
      artists.append("아티스트 정보없음")

    #세부정보
    if driver.find_element_by_css_selector("dl.list"):
      detail_box = driver.find_element_by_css_selector("dl.list")

      if detail_box.find_elements_by_css_selector("dd"):
        detail_dd = detail_box.find_elements_by_css_selector("dd")

        if detail_dd[0]:
          albums.append(detail_dd[0].text)
        else:
          albums.append("앨범없음")

        if detail_dd[1]:
          dates.append(detail_dd[1].text)
        else:
          dates.append("발매일 없음")

        if detail_dd[2]:
          genres.append(detail_dd[2].text)
        else:
          genres.append("장르없음")

        try:
          if detail_dd[3]:
            flacs.append(detail_dd[3].text)
        except:
          flacs.append("flac 없음")   
      else:
        pass
    else:
      pass

    #가사
    if driver.find_element_by_css_selector("div#lyricArea"):
      try:
        cont = driver.find_element_by_css_selector("div#d_video_summary")
        lyrics.append(cont.text.strip())
      except:
        lyrics.append("가사없음")


    driver.back()
    time.sleep(1.5)

    song_info_list = driver.find_elements_by_css_selector("a.btn.btn_icon_detail") #50개
    
    
frame = pd.DataFrame({"title": titles,
                      "artist": artists,
                      "album": albums,
                      "date": dates,
                      "genre": genres,
                      "flac": flacs,
                      "lyric": lyrics})

print("9페이지까지 개수:", len(frame))
frame
```
